```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from pathlib import Path


file_path = Path('/Users/terahskillings/Desktop/HuluRaw.csv')


if file_path.is_file():
    print(f"The file {file_path} exists.")
    
    df = pd.read_csv(file_path)
    print(df.head()) 
else:
    print(f"The file {file_path} does not exist.")
```

```python
genre_counts = df['show/genre'].value_counts()
top_5_genres = genre_counts.head(5)

fig, ax = plt.subplots(figsize=(12, 6))
ax.barh(top_5_genres.index, top_5_genres.values, color='blue')
ax.set_title('Top 5 Genres', fontsize=16)
ax.set_xlabel('Count', fontsize=14)
ax.set_ylabel('Show/Genre', fontsize=14)
ax.tick_params(axis='both', labelsize=12)  

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.hist(df['show/episodes_count'], bins=25, alpha=0.7, color='blue', edgecolor='black')
ax.set_title('Histogram of Episodes', fontsize=16)
ax.set_xlabel('Episode Count', fontsize=14)
ax.set_ylabel('Frequency', fontsize=14)
ax.tick_params(axis='both', labelsize=12) 
plt.show()

```

```python
comedy_df = df[df['show/genre'] == 'Comedy']
rating_count = comedy_df['show/show_rollups/showtime/highest_rating'].value_counts()

fig, ax = plt.subplots(figsize=(12, 10))
ax.bar(rating_count.index, rating_count.values, color='blue')
ax.set_title('Comedy Ratings', fontsize=16)
ax.set_xlabel('Rating Type', fontsize=14)
ax.set_ylabel('Count', fontsize=16)
ax.tick_params(axis='both', labelsize=16)  
plt.show()
```
