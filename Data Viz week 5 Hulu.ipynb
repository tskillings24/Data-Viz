```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from pathlib import Path


file_path = Path('/Users/terahskillings/Desktop/HuluRaw.csv')


if file_path.is_file():
    print(f"The file {file_path} exists.")
    
    df = pd.read_csv(file_path)
    print(df.head()) 
else:
    print(f"The file {file_path} does not exist.")
```

```python
df.describe()
```

```python
df
```

```python
genre_counts = df['show/genre'].value_counts()
top_5_genres = genre_counts.head(5)

plt.figure(figsize=(12, 6))
plt.bar(top_5_genres.index, top_5_genres.values)
plt.title('Top 10 Genres', fontsize=16)
plt.xlabel('Show/Genre', fontsize=14)
plt.ylabel('Count', fontsize=14)

plt.show()
```

```python
plt.hist(df['show/episodes_count'], bins=25, alpha=0.7, color='blue', edgecolor='black') 
plt.title('Histogram of Episodes')
plt.xlabel('Episode Count')
plt.ylabel('Frequency')
plt.show()
```

```python
comedy_df = df[df['show/genre'] == 'Comedy']
```

```python
rating_count = comedy_df['show/show_rollups/showtime/highest_rating'].value_counts()
plt.figure(figsize=(12, 6))
plt.bar(rating_count.index, rating_count.values)
plt.title('Comedy Ratings', fontsize=16)
plt.xlabel('Rating Type', fontsize=14)
plt.ylabel('Count', fontsize=14)

plt.show()
```
